```python
import feather
import numpy as np
import pandas as pd 
```

```python
np.random.seed=56
df_size= 10_000_000
df = pd.DataFrame({
    'a': np.random.rand(df_size),
    'b': np.random.rand(df_size),
    'c': np.random.rand(df_size),
    'd': np.random.rand(df_size),
    'e': np.random.rand(df_size)
})
df.head()
```

```python
df 
```

```python
# save feather from pandas
df.to_feather('1M.feather')
```

```python
# read feather file from pandas
df= feather.read_dataframe('1M.feather')
```
